```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output
import time
```

```python
class Robot:

    # Definamos el constructor de clase
    def __init__(self, f, rate, Id = 0):

      # Inicializar atributos
      self.Id = Id
      self.f = f
      self.rate = rate

      # Necesitamos ubicar al robot
      self.r = np.random.uniform(-10,10)

    # Metodo de clase
    def Mutate(self):
      self.r += np.random.normal(loc=0,scale=self.rate)

    def __repr__(self):
      return f"Robot {self.Id}"
```

```python
f = lambda x: x**2 + 10*np.sin(x)
```

```python
rate = 0.05
```

```python
def GetRobots(N=10):

  Robots = []
  for i in range(N):
    r = Robot(f,rate,Id=i)
    Robots.append(r)

  return Robots
```

```python
Robots = GetRobots(20)
```

```python
def Genetic(Robots, epochs=100):

  t = np.linspace(-5,5,50)

  for e in range(epochs):

    clear_output(wait=True)

    # Primero mutamos
    for i, p in enumerate(Robots):
      p.Mutate()

    for i in range(len(Robots)):
      plt.scatter(Robots[i].r,Robots[i].f(Robots[i].r),color='r')

    plt.show()
    time.sleep(0.1)
```

```python
Genetic(Robots)
```
