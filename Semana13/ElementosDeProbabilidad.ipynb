```python
def GetExperiment(N = 1000):

  freq = np.zeros(N)

  for i in range(int(N)):
    d1 = np.random.randint(1,7)
    d2 = np.random.randint(1,7)

    freq[i] = d1 + d2

  return freq
```

```python
freq = GetExperiment()
```

```python
plt.hist(freq)
```

```python
prob7 = np.sum( freq == 7 ) / len(freq)
prob7
```

```python
prob2 = np.sum( freq == 2 ) / len(freq)
prob10 = np.sum( freq == 10 ) / len(freq)
print(prob2 + prob10)
```
