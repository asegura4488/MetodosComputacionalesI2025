```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def GetFrequencies(n):

    NTimes = 0

    for i in range(int(n)):
        t = np.random.randint(1,7)
        if t == 2:
            NTimes += 1

    return NTimes/n
```

```python
n = np.logspace(1,6,10)
freqr = np.zeros(len(n))
for i in range(len(n)):
    freqr[i] = GetFrequencies(n[i])
```

```python
plt.plot(n,freqr)
plt.axhline(y=1/6,color='r',ls='--')
plt.xscale('log')
plt.ylabel(r'$freq_{rel}$')
```

```python
def GetExperiment(N = 1000):

  freq = np.zeros(N)

  for i in range(int(N)):
    d1 = np.random.randint(1,7)
    d2 = np.random.randint(1,7)

    freq[i] = d1 + d2

  return freq
```

```python
freq = GetExperiment()
```

```python
# por favor no hacer histogramas de esta forma
plt.hist(freq)
```

```python
# Vamos a construir el objeto histograma
# Definimos las categorias del histograma
x = np.linspace(2,13,12)
x
```

```python
Rango = x[-1] - x[0]
Rango
```

```python
h, bins = np.histogram(freq, bins=x)
# Necesitamos el ancho
w = np.diff(bins)
# Necesitamos la marca de clase
m = bins[:-1] #+ 0.5*w[0]
m
```

```python
# Vamos a pintar
plt.bar(m,h,width=w,ec='k')
```

```python
# Necesitamos normalizar a 1
I = np.sum(w*h)
I
```

```python
hnorm = h/I
hnorm
```

```python
np.sum(w*hnorm)
```

```python
plt.bar(m,hnorm,width=w,ec='k')
```

```python
# Valor medio
X = bins[:-1]
mean = np.sum( hnorm * X )
mean
```

```python
varianza = np.sum( (X - mean)**2* hnorm )
varianza
```

```python
std = np.sqrt(varianza)
std
```

```python
# Vamos a calcular el intervalo de confiaza 68%
IC = [mean - std, mean + std]
IC
```
