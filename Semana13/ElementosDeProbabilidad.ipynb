```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def GetFrequencies(n):

    NTimes = 0

    for i in range(int(n)):
        t = np.random.randint(1,7)
        if t == 2:
            NTimes += 1

    return NTimes/n
```

```python
n = np.logspace(1,6,10)
freqr = np.zeros(len(n))
for i in range(len(n)):
    freqr[i] = GetFrequencies(n[i])
```

```python
plt.plot(n,freqr)
plt.axhline(y=1/6,color='r',ls='--')
plt.xscale('log')
plt.ylabel(r'$freq_{rel}$')
```

```python
def GetExperiment(N = 1000):

  freq = np.zeros(N)

  for i in range(int(N)):
    d1 = np.random.randint(1,7)
    d2 = np.random.randint(1,7)

    freq[i] = d1 + d2

  return freq
```

```python
freq = GetExperiment()
```

```python
# por favor no hacer histogramas de esta forma
plt.hist(freq)
```

```python
# Vamos a pintar
plt.bar(m,h,width=w,ec='k')
```
