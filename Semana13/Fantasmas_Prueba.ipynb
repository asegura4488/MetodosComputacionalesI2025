```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
from scipy.special import factorial
```

```python
k = np.array([0,1,2,3,4,5,6,7])
```

```python
def Poisson(k,lambda_=2):

  return np.exp(-lambda_)*lambda_**k / factorial(k)
```

```python
p = Poisson(k)
```

```python
plt.stem(k,p)
```
