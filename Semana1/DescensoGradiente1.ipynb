```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output
import time
import copy
```

```python
def Plotter(e):

  fig = plt.figure(figsize=(8,3))
  ax =fig.add_subplot(121)
  ax1 =fig.add_subplot(122)
  ax.set_title('{:.0f}'.format(e),fontsize=10)

  return ax,ax1
```

```python
Plotter(10)
```

```python
def f(x):
  return x**2 + 10*np.sin(x)
```

```python
xvalues = np.linspace(-10,10,50)
yvalues = f(xvalues)
```

```python
def Derivative(f,x,h=1e-5):
  return (f(x+h)-f(x-h))/(2*h)
```

```python
plt.plot(xvalues,yvalues)
```

```python
def GradientDescent(f,epochs=100,lr=1):

  # Semilla
  xn = np.random.randint(-20,20)
  History = np.array([f(xn)])

  for e in range(epochs):

    # Algoritmo
    xn1 = xn - lr*Derivative(f,xn) # MEnos es todo!!!

    xn = xn1

    History = np.append(History,f(xn))

    clear_output(wait=True)
    ax,ax1 = Plotter(e)
    ax.plot(xvalues,yvalues)
    ax.scatter(xn,f(xn),color='red')
    ax1.plot(History[:e])

    time.sleep(0.2)
    plt.show()

  print(xn)
```

```python
GradientDescent(f,epochs=200,lr=0.05)
```
