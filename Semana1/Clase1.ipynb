```python
# Ejemplo 2
lista2 = [3,2,10,15,7]
import numpy as np
l = np.sort(lista2)
l
```

```python
# algoritmo para ordenar
for i in range(len(lista2)):
  for j in range(len(lista2)):
    if lista2[i] < lista2[j]:
      aux = lista2[i]
      lista2[i] = lista2[j]
      lista2[j] = aux
#
```
