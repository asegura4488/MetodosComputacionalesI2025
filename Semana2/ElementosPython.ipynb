```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def AproximacionPi(r = 1., niter = 10):

  print(r,niter)

  resultados = [] # Nuestra estructura de datos
  for k in range(niter):


    # Definamos el poligono
    n = 6 * 2 ** k

    # Calculando los lados
    Lin = 2 * r * np.sin( np.pi/ n )
    Lout = 2 * r * np.tan( np.pi/ n )
    # Calculandos los perimetros
    Pin = n * Lin
    Pout = n * Lout
    # Calculamos pi
    pi_in = Pin / (2*r)
    pi_out = Pout / (2*r)

    resultados.append({
        "k": k,
        "pi_in": pi_in,
        "pi_out": pi_out
    })

    print(k,n,Lin,Lout,Pin,Pout,pi_in,pi_out)

  return resultados
```

```python
resultados = AproximacionPi(r=1, niter=15)
```

```python
resultados
```

```python
k = [iterador['k'] for iterador in resultados]
pi_in = [row['pi_in'] for row in resultados]
pi_out = [row['pi_out'] for row in resultados]
```
