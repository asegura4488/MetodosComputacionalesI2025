```python
def Plotter():

  fig = plt.figure(figsize=(8,3))
  ax =fig.add_subplot(121)
  ax1 =fig.add_subplot(122)
  #ax2 =fig.add_subplot(312)

  return ax,ax1
```

```python
Plotter()
```

```python
# Programamos el algoritmo
def Babilonico(x,f,i):

  if i == 0:
    f = x
  else:
    f = (f+x/f)/2
  return f
```

```python
# Vamos a animar
def RaizCuadrada(x,epochs):

  sucesion = np.array([x])
  distancias = np.array([])


  for e in range(1,epochs):

    clear_output(wait=True)
    ax,ax1 = Plotter()
    ax.set_title("{:.0f}".format(e))

    # Calculamos la raiz y cuadrada y llenamos
    raiz = Babilonico(x,sucesion[e-1],e)
    sucesion = np.append(sucesion,raiz)

    presente = e
    pasado = e-1
    distancias = np.append(distancias,abs(sucesion[presente]-sucesion[pasado]))



    # Graficamos
    ax.plot(sucesion[:e],'.-',color='k')
    ax1.plot(distancias,'.-',color='k')

    plt.show()

    if distancias[-1] < 1e-16:
      break

    #print(e,raiz)

    time.sleep(0.22)

    raiz = Decimal(raiz)
    #print(f"{e:.0f}",f"{raiz:.100f}")

  return sucesion
```

```python
RaizCuadrada(2,20)
```
