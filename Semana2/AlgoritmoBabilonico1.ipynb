```python
import numpy as np
import matplotlib.pyplot as plt
import time
from IPython.display import clear_output
from decimal import Decimal, getcontext
#%matplotlib inline
```

```python
getcontext().prec = 100
```

```python
Decimal(2.)
```

```python
def Plotter():

  fig = plt.figure(figsize=(8,3))
  ax =fig.add_subplot(121)
  ax1 =fig.add_subplot(122)
  #ax2 =fig.add_subplot(312)

  return ax,ax1
```

```python
Plotter()
```

```python
# Programamos el algoritmo
def Babilonico(x,f,i):

  if i == 0:
    f = x
  else:
    f = (f+x/f)/2
  return f
```

```python
Babilonico(2,1.4166666666666665,3)
```

```python
# Vamos a animar
def RaizCuadrada(x,epochs):

  sucesion = np.array([x])
  distancias = np.array([])


  for e in range(1,epochs):

    clear_output(wait=True)
    ax,ax1 = Plotter()
    ax.set_title("{:.0f}".format(e))

    # Calculamos la raiz y cuadrada y llenamos
    raiz = Babilonico(x,sucesion[e-1],e)
    sucesion = np.append(sucesion,raiz)

    presente = e
    pasado = e-1
    distancias = np.append(distancias,abs(sucesion[presente]-sucesion[pasado]))

    #print(distancias)


    # Graficamos
    ax.plot(sucesion[:e],'.-',color='k')


    plt.show()

    if distancias[-1] < 1e-34:
      break

    #print(e,raiz)

    time.sleep(0.01)

    raiz = Decimal(raiz)
```

```python
RaizCuadrada(2,100)
```
