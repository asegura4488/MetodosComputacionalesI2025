```python
# Programamos el algoritmo
def Babilonico(x,f,i):

  if i == 0:
    f = x
  else:
    f = (f+x/f)/2
  return f
```

```python
# Vamos a animar
def RaizCuadrada(x,epochs):

  sucesion = np.array([x])


  for e in range(1,epochs):

    clear_output(wait=True)
    #ax,ax1 = Plotter()
    #ax.set_title("{:.0f}".format(e))

    # Calculamos la raiz y cuadrada y llenamos
    raiz = Babilonico(x,sucesion[e-1],e)
    sucesion = np.append(sucesion,raiz)

    # Graficamos
    plt.plot(sucesion)

    #plt.show()
    print(e,raiz)

    time.sleep(1)
```

```python
RaizCuadrada(2,10)
```
