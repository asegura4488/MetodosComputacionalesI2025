```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Recta(x):
  return 3*x + 2
```

```python
p = np.array([5,4])
```

```python
x = np.linspace(0,5,10)
plt.plot(x,Recta(x))
plt.scatter(p[0],p[1],color='r')
```

```python
def GetDistance(x):
  return np.sqrt((x-p[0])**2 + (Recta(x)-p[1])**2)
```

```python
GetDistance(1)
```

```python
x = np.linspace(0,3,30)
d = GetDistance(x)
plt.scatter(x,d)
```

```python
def Derivative(f,x,h=1e-6):
  return (f(x+h)-f(x-h))/(2*h)
```

```python
DR = Derivative(Recta,x)
```
