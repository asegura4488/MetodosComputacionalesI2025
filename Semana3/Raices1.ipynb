```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Recta(x):
  return 3*x + 2
```

```python
p = np.array([5,4])
```

```python
x = np.linspace(0,5,10)
plt.plot(x,Recta(x))
plt.scatter(p[0],p[1],color='r')
```

```python
def GetDistance(x):
  return np.sqrt((x-p[0])**2 + (Recta(x)-p[1])**2)
```

```python
GetDistance(1)
```

```python
x = np.linspace(0,3,100)
d = GetDistance(x)
plt.scatter(x,d)
```

```python
def Derivative(f,x,h=1e-5):
  return (f(x+h)-f(x-h))/(2*h)

def Derivative2(f,x,h=1e-5):
  return (f(x+h)-2*f(x)+f(x-h))/(h**2)
```

```python
DR = Derivative(GetDistance,x)
```

```python
plt.plot(x,DR)
plt.axhline(y=0,color='r')
plt.grid()
```

```python
minimo = 0
for i in range(len(x)-1):
  fi = Derivative(GetDistance,x[i])
  fj = Derivative(GetDistance,x[i+1])
  if fi*fj < 0:
    print(x[i],GetDistance,x[i+1])
    minimo = x[i]
    break
```

```python
def GetNewtonMethod(f,xn,itmax=100,precision=1e-10):

  error = 1
  it = 0

  while error > precision and it < itmax:

    try:
      xn1 = xn - Derivative(f,xn)/Derivative2(f,xn)
      print(xn1,Derivative(f,xn),Derivative2(f,xn))
      error = np.abs(Derivative(f,xn)/Derivative2(f,xn))
    except ZeroDivisionError:
      print('Division por cero')

    xn = xn1
    it += 1

  if it == itmax:
    False
  else:
    print(it)
    return xn
```
