```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def function(x):

  #return 0.5*(5*x**3 - 3*x)
  return (35*x**4 - 30*x**2 + 3)/8
  #return (429*x**7 - 693*x**5 + 315*x**3 - 35*x) / 16
  #return x**2 + 1
```

```python
x = np.linspace(-1,1,100)
y = function(x)
```

```python
plt.plot(x,y)
plt.axhline(y=0,color='r',linestyle='-')
```

```python
def Derivative(f,x,h=1e-6):
  return (f(x+h)-f(x-h))/(2*h)
```

```python
def GetNewtonMethod(f,df,xn,itmax=100,precision=1e-12):

  error = 1
  it = 0

  sucesion = np.array([xn])

  while error > precision and it < itmax:

    try:
      dfxn = df(f,xn)
      if np.abs(dfxn) < 1e-16:
        return False

      xn1 = xn - f(xn)/dfxn
      error = np.abs(xn1-xn)

    except ZeroDivisionError:
      return False

    sucesion = np.append(sucesion,xn1)
    xn = xn1
    it += 1

  if it == itmax:
    return False,sucesion
  else:
    print(it)
    return xn,sucesion
```

```python
semillas = np.linspace(-1,1,10)
for i in semillas:
  root,sucesion = GetNewtonMethod(function,Derivative,i)
  print(root)
  plt.scatter(np.arange(0,len(sucesion),1),sucesion)
  plt.axhline(y=root,color='r')
```
