```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definir es el conjunto de soporte
x = np.linspace(0,2*np.pi,10)
#x
```

```python
def funcion(x):
  return np.sin(x)
```

```python
y = funcion(x)
#y
```

```python
plt.plot(x,y,'.')
```

```python
def ExactDerivada(x):
  return np.cos(x)
```

```python
h = x[1]-x[0]
h
```

```python
def RightDerivative(f,x,h):

  d = 0.
  if h != 0:
    d = (f(x+h) - f(x))/h

  return d
```

```python
RightDerivative(funcion,np.pi,0.4)
```

```python
RDerivative = RightDerivative(funcion,x,h)
```

```python
plt.plot(x,RDerivative,'.',label='Derecha')
plt.plot(x,np.cos(x),'.',label='Exact')
plt.legend()
```
