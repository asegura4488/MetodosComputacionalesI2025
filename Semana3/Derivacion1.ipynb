```python
def ExactDerivada(x):
  return np.cos(x)
```

```python
h = x[1]-x[0]
h
```

```python
def RightDerivative(f,x,h):

  d = 0.
  if h != 0:
    d = (f(x+h) - f(x))/h

  return d
```

```python
def LeftDerivative(f,x,h):

  d = 0.
  if h != 0:
    d = (f(x) - f(x-h))/h

  return d
```

```python
def CentralDerivative(f,x,h):

  d = 0.
  if h != 0:
    d = (f(x+h) - f(x-h))/(2*h)

  return d
```

```python
def SusanDerivative(f,x,h):

  d = 0.
  if h != 0:
    d = (-f(x+2*h) + 6*f(x+h) -3*f(x) - 2*f(x-h))/(6*h)

  return d
```
