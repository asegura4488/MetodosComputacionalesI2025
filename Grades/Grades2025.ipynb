```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Files = ['Metodos1.xlsx']
df = pd.read_excel(Files[0])
df = df.fillna(0)
df.head(20)
```

```python
df.keys()
```

```python
dfparciales = df[['Parcial 1','Parcial 2','Parcial 3', 'Parcial 4']]
dftalleres = df[['Tarea 1', 'Tarea 2', 'Tarea 3', 'Tarea 4']]
```

```python
# Porcentaje talleres
df = df.assign( Talleres = dftalleres.mean(axis=1)*0.15 )
# Porcentaje parciales
df = df.assign( Parciales = dfparciales['Parcial 1']*0.20 + \
                            dfparciales['Parcial 2']*0.25 + \
                            dfparciales['Parcial 3']*0.25 + \
                            dfparciales['Parcial 4']*0.15
              )
```

```python
df[['Talleres','Parciales']]
```

```python
# ponderación acordada
df = df.assign( NotaFinal = np.round(0.75*(df.Parciales+df.Talleres) + 0.25*df.Complementaria,2) ) 
#df
```

```python
dffinal = df[['Last Name','First Name', 'NotaFinal']]
dffinal
#dffinal.sort_values(by='NotaFinal')
```

```python
dffinal.to_excel('Metodos1Final.xlsx',index=False,float_format="%.2f")
```

```python
percentiles = np.percentile(dffinal.NotaFinal,[0,25,50,75,100])



fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(111)

ax.set_xticklabels(['Métodos computacionales 1'], rotation=1, fontsize=15)

ax.boxplot(dffinal.NotaFinal)
ax.set_ylabel('Grades', fontsize=15)

for i in range(len(percentiles)):
    ax.text(1+0.1,percentiles[i],np.round(percentiles[i],2),color='r')
```
