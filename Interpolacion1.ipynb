```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

```python
X = np.array([0,1,-0.5,-2])
Y = np.array([0,2,3,-1])
```

```python
plt.scatter(X,Y)
```

```python
# Queremos crear las bases cardinales sobre el soporte
def Lagrange(x,X,i):

  L = 1
  for j in range(len(X)):
    if i != j:
      L *= (x - X[j])/(X[i]-X[j])

  return L
```

```python
x = sym.Symbol('x',real=True)
Lagrange(x,X,0)
```

```python
Lagrange(x,X,1).simplify()
```

```python
Lagrange(x,X,0).subs(x,X[0])
```

```python
Lagrange(x,X,1).subs(x,X[2])
```
