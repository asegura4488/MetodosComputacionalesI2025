```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

```python

X = np.array([-2,-0.5,0,1,2,3])
Y = np.array([-1,3,0,2,-10,5])
```

```python
plt.scatter(X,Y)
```

```python
# Queremos crear las bases cardinales sobre el soporte
def Lagrange(x,X,i):

  L = 1
  for j in range(len(X)):
    if i != j:
      L *= (x - X[j])/(X[i]-X[j])

  return L
```

```python
x = sym.Symbol('x',real=True)
Lagrange(x,X,0)
```

```python
Lagrange(x,X,1).simplify()
```

```python
Lagrange(x,X,0).subs(x,X[0])
```

```python
Lagrange(x,X,1).subs(x,X[2])
```

```python
def Interpolate(x,X,Y):

  Poly = 0
  for i in range(len(X)):
    Poly += Y[i]*Lagrange(x,X,i)

  return Poly
```

```python
Interpolate(x,X,Y).simplify()
```

```python
plt.scatter(X,Y,color='r')
x1 = np.linspace(X[0],X[-1],50)
#x1 = np.linspace(-4,4,50)

print(x1)
y1 = Interpolate(x1,X,Y)
plt.plot(x1,y1)
```

```python
poly = Interpolate(x,X,Y).simplify()
poly.subs(x,-0.25)
```
